*1)  What player on a team with “ia” in the team name played less than 200 minutes and made more than 100 passes? Return the player surname.*

```python
%%sql
select * from Players where minutes < 200 and passes > 100 and team like "%ia%";
```

*2) Find all players who made more than 20 shots. Return all player information in descending order of shots made.*

*3) Find the goalkeepers of teams that played more than four games. List the surname of the goalkeeper, the team, and the number of minutes the goalkeeper played.*

```python
%%sql
select Players.surname, Players.team, Players.minutes 
from Players inner join Teams on Players.team = Teams.team 
where Teams.games > 4 and Players.position = 'goalkeeper';
```

*4) How many players who play on a team with ranking <10 played more than 350 minutes? Return one number in a column named 'superstar'.*

```python
%%sql
SELECT Players.team, COUNT(Players.surname) AS 'superstar'
FROM Players
INNER JOIN Teams ON Players.team = Teams.team 
WHERE Teams.ranking < 10 AND Players.minutes > 350
GROUP BY Players.team;
```

*5) What is the average number of passes made by forwards? By midfielders? Write one query that gives both values with the corresponding position.*

```python
%%sql
SELECT position, ROUND(AVG(passes))
FROM Players
WHERE position = 'forward' OR position = 'midfielder'
GROUP BY position;
```

*6) Which team has the highest average number of passes per minute played? Return the team and average passes per minute. Hints: (1) You can compute a team's average number of passes per minute played by dividing the total number of passes by the total number of minutes. To force floating point division, multiply one operand by 1.0. (2) Consider using Limit.*

```python
%%sql
SELECT team, ROUND(CAST(SUM(passes) AS FLOAT) / CAST(SUM(minutes) AS FLOAT), 5) AS AVG_PPM
FROM Players
GROUP BY team;
```

#### <font color = 'green'>Titanic Basic Queries</font>

*1) How many married women over age 50 embarked in Cherbourg? Note: To check if attribute A begins with string S use "A like 'S%'"*

```python
%%sql
SELECT last, first
FROM Titanic
WHERE age > 50 AND embarked = 'Cherbourg' AND first LIKE 'Mrs%'
```

*2) List the average fare paid by passengers in each of the embarkation cities (along with the city), in descending order of average fare.*

```python
%%sql
SELECT embarked, ROUND(AVG(fare), 3) as 'Average fare'
FROM Titanic
GROUP BY embarked
ORDER BY 'Average fare' DESC
```

#### <font color = 'green'>Titanic Advanced Queries</font>

*1) Are there any pairs of passengers with the same last name where one is in first class and the other is in third class? If so, return the last name and the two first names. Label the first-name columns 'first' for the passenger in first class and 'third' for the passenger in third class.*

```python
%%sql
SELECT T1.last, T1.first, T2.first AS second
FROM Titanic AS T1, Titanic AS T2
WHERE T1.class = 1 AND T2.class = 3 AND T1.last = T2.last;
```

*2) Which embarkation cities have more than 40 passengers whose age is missing? Reminder: Blanks in SQL tables are given a special value called 'null', and conditions 'A is null' and 'A is not null' can be used in Where clauses to check whether attribute A has the 'null' value.*

```python
%%sql
SELECT embarked, COUNT(last) AS passengers
FROM Titanic
WHERE age IS NULL
GROUP BY embarked
HAVING passengers > 40;
```

*3) Find all classes where the average fare paid by passengers in that class was more than twice the overall average or less than half the overall average.*

```python
%%sql
SELECT T1.class, AVG(T1.fare) AS avg_fare, T2.avg_all
FROM Titanic AS T1
INNER JOIN (SELECT AVG(fare) AS avg_all FROM Titanic) AS T2
GROUP BY T1.class
HAVING avg_fare > 2 * T2.avg_all OR avg_fare < T2.avg_all / 2;
```
