### <font color = 'green'>World Cup with Pandas</font>
Solve the problems using the data in Pandas dataframes.

```python
# Read the Players and Teams data into dataframes
f = open('Players.csv')
players = pd.read_csv(f)
f = open('Teams.csv')
teams = pd.read_csv(f)
```

```python
# What player on a team with “ia” in the team name played less than
# 200 minutes and made more than 100 passes? Print the player surname.
players[(players['minutes'] < 200) & (players['passes'] > 100) & (players['team'].str.contains('ia'))]
```

```python
# What is the average number of passes made by forwards? By midfielders?
# Make sure the answer specifies which is which, and don't include other
# positions in your result.
players.loc[:, ['position', 'passes']].groupby('position').mean().loc[['forward', 'midfielder'],]
```

```python
# Which team has the highest ratio of goalsFor to goalsAgainst?
# Print the team name only.
# Hint: Add a "ratio" column to the teams dataframe, then sort,
# then use head(1) or tail(1) depending how you sorted
teams.loc[(teams['goalsFor'] / teams['goalsAgainst']).idxmax(), 'team']
```

```python
# How many players who play on a team with ranking <10 played
# more than 350 minutes?
# Reminder: len() gives number of rows in a dataframe
top_9_team = teams[teams['ranking'] < 10]['team'].values.tolist()
top_players = players.set_index('team').loc[top_9_team, ['minutes']]
top_players = top_players[top_players['minutes'] > 350].groupby('team').count().rename(columns={'minutes': 'Total'})
top_players
```

### <font color = 'green'>World Cup Data Visualization</font>


```python
# Re-read Players.csv and Teams.csv into lists of dictionaries
# Read the Players and Teams data into dataframes
f = open('Players.csv')
players = pd.read_csv(f)
f = open('Teams.csv')
teams = pd.read_csv(f)
```

```python
# Create a scatterplot of players showing passes made (y-axis)
# versus minutes played (x-axis). Color each player based on their
# position (goalkeeper, defender, midfielder, forward).
position = ['goalkeeper', 'defender', 'midfielder', 'forward']
color = ['red', 'blue', 'green', 'black']
po_color = dict(zip(position, color))

patchList = []
for key in po_color:
    data_key = mpatches.Patch(color = po_color[key], label = key)
    patchList.append(data_key)


plt.figure(figsize = (10, 10))
plt.title('Relationship between passes and minutes played')
plt.scatter(x = players['minutes'], y = players['passes'], color = players['position'].apply(lambda x: po_color[x]))
plt.legend(handles = patchList)
```

```python
# Create a bar chart showing the number of teams with 0, 1, and 2 red cards.
bars = ['zero', 'one', 'two'] # number of red cards
heights = [0, 0, 0] # number of teams with zero/one/two red cards

heights = teams[['team', 'redCards']].groupby(['redCards']).count().rename(columns={'team': 'No of teams'}).values.flatten().tolist()

plt.figure(figsize = (10, 10))
plt.title('Number of teams with 0, 1, 2 red cards')
plt.bar(x = bars, height = heights, color = 'red')
```

```python
# Create a pie chart showing the relative percentage of teams
# with 0, 1, and 2 red cards.
plt.figure(figsize = (10, 10))
plt.title('Percentage of teams with 0, 1, 2 red cards')
plt.pie(heights, labels = bars)
```

### <font color = 'green'>Titanic Data Visualization</font>


```python
# Read the Titanic data into a dataframe
f = open('Titanic.csv')
titanic = pd.read_csv(f)
titanic.head()
```

```python
# Create a bar chart showing the average fare paid by passengers in each class.
# The three bars should be labeled 'first', 'second', 'third'.
class_type = ['first', 'second', 'third']
fares = titanic[['class', 'fare']].groupby(['class']).mean().values.flatten().tolist()

plt.figure(figsize = (10, 10))
plt.title('Percentage of average fare in 3 classes')
plt.bar(x = class_type, height = fares)
```

```python
# Create a pie chart showing the relative number of male survivors, male non-survivors,
# female survivors, and female non-survivors (four slices).

num = titanic[['gender', 'survived', 'first']].groupby(['gender', 'survived']).count().rename(columns = {'first': 'count'})

plt.figure(figsize = (10, 10))
plt.title('Relative number of survivors and gender')
plt.pie(x = num.values, labels = ['Female non-survivors', 'Female survivors', 'Male non-survivors', 'Male survivors'])
```
